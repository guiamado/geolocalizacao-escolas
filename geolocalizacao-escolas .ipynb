```python
import requests
from time import sleep
import pandas as pd
import folium
```

```python
enderecos = pd.read_csv('enderecos.csv')
df_escolas = enderecos.head(10)
df_escolas
```

```python
GOOGLE_MAPS_API_URL = 'https://nominatim.openstreetmap.org/search.php?format=json'
```

```python
params = {
    'q': '',
}
```

```python
df_school = enderecos.head(120)
df_school

for index, row in df_school.iterrows():
    params['q'] = row['escola']
    req = requests.get(GOOGLE_MAPS_API_URL, params=params)
    res = req.json()
    try:
        if len(res) > 0:
            print(res[0]['lat'], res[0]['lon'])
            df_school.loc[index,'lat'] = res[0]['lat']
            df_school.loc[index,'long'] = res[0]['lon']
        if len(res) < 1:
            params['q'] = row['endereco_bairro'] + ' ' +row['endereco_logradouro'] + row['endereco_complemento']
            req = requests.get(GOOGLE_MAPS_API_URL, params=params)
            res = req.json()
            print(res[0]['lat'], res[0]['lon'])
            df_school.loc[index,'lat'] = res[0]['lat']
            df_school.loc[index,'long'] = res[0]['lon']
    except:
        print(res)
        df_school.loc[index,'lat'] = -16.680882
        df_school.loc[index,'long'] = -49.2532691
    sleep(0.5)

df_school
```

```python
escolares = folium.Map(
    location=[-15.7801, -47.9292],
    zoom_start=7
)

from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(escolares)

for _, row in df_school.iterrows():
#     if ubs['lat'] != 0 and ubs['long'] != 0:
    como_chegar = 'https://www.google.com/maps/dir/?api=1&origin=current+location&destination='\
    + str(row['lat']) + ',' + str(row['long'])
    
    html = '<h5 style="margin-bottom:6px">' + row['escola'] + '</h5>' \
    + 'Tipo da Escola: ' +str(row['tipo_escola'])\
    + '<br> Região Administrativa: ' +str(row['endereco_regiao_administrativa'])\
    + '<br> Endereço: ' +str(row['endereco_logradouro']) +str(row['endereco_complemento'])\
    + '<br> <a target="_blank" href="'+como_chegar+'">Como Chegar</a>'
    iframe = folium.IFrame(html=html, width=250, height=150)
    popup = folium.Popup(iframe, max_width=2650)
    
    marker = folium.Marker(
        location=[float(row['lat']), float(row['long'])],
        popup=popup,
        icon=folium.Icon(color='red')
    ).add_to(marker_cluster)
    
escolares
```
